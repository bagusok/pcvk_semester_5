```
import numpy as numpy
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io
from skimage import transform
from PIL import Image
import matplotlib.pyplot as plt
```

```
urls = [
    "https://iiif.lib.ncsu.edu/iiif/0052574/full/800,/0/default.jpg",
    "https://iiif.lib.ncsu.edu/iiif/0016007/full/800,/0/default.jpg",
    "https://freepngimg.com/download/temp_jpeg/22193-3-adorable-cat.jpeg"
]

for url in urls:
    image = io.imread(url)
    image = cv.resize(image, (0, 0), fx=0.5, fy=0.5)
    image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

    final_frame = cv.hconcat((image, image_2))
    cv2_imshow(final_frame)
    print('\n')
```

## Kegunaan kode [255,255,255]? Jelaskan!
Kode [255,255,255] merepresentasikan warna putih dalam ruang warna RGB (Red, Green, Blue). Dalam model warna RGB, setiap warna dibentuk dengan menggabungkan intensitas warna merah, hijau, dan biru. Nilai intensitas berkisar dari 0 hingga 255.

[255, 0, 0] adalah merah murni.
[0, 255, 0] adalah hijau murni.
[0, 0, 255] adalah biru murni.
[0, 0, 0] adalah hitam.
[255, 255, 255] adalah putih, yang merupakan kombinasi intensitas maksimum dari merah, hijau, dan biru.
Dalam konteks kode Anda di sel qipcUHS1xSWz, image_3[int((tinggi)/2), y] = [255,255,255] digunakan untuk mengubah warna piksel pada baris tengah gambar (int((tinggi)/2)) untuk setiap kolom (y) menjadi putih. Ini akan menghasilkan garis horizontal putih di tengah gambar.

## Keterkaitan antara pixel dan juga resolusi gambar yang tinggi ataupun rendah!

Pixel adalah elemen terkecil dari sebuah gambar digital. Setiap piksel adalah sebuah titik kecil yang memiliki warna tertentu. Gambar digital dibentuk oleh grid atau matriks piksel-piksel ini.
Resolusi gambar mengacu pada jumlah piksel yang membentuk gambar, biasanya dinyatakan sebagai lebar dikalikan tinggi (misalnya, 1920x1080 piksel).
Keterkaitannya adalah:

Resolusi tinggi berarti gambar terdiri dari jumlah piksel yang lebih banyak. Dengan lebih banyak piksel dalam area yang sama, gambar dapat menampilkan detail yang lebih halus dan tajam. Gambar resolusi tinggi biasanya memiliki ukuran file yang lebih besar dan membutuhkan lebih banyak memori untuk diproses.
Resolusi rendah berarti gambar terdiri dari jumlah piksel yang lebih sedikit. Ini menghasilkan gambar dengan detail yang kurang tajam dan mungkin terlihat "pecah" atau berpiksel (pixelated) ketika diperbesar. Gambar resolusi rendah memiliki ukuran file yang lebih kecil dan lebih cepat untuk diproses.
Jadi, resolusi gambar secara langsung menentukan seberapa banyak piksel yang ada dalam gambar, dan jumlah piksel ini pada gilirannya memengaruhi tingkat detail dan kualitas visual gambar tersebut. Semakin tinggi resolusinya, semakin banyak pikselnya, dan semakin detail gambarnya (dengan asumsi kualitas pengambilan gambar yang baik).

```
# Buat Garis Vertikal
img_4 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

tinggi_3 = img_4.shape[0]
lebar_3 = img_4.shape[1]

for j in range(tinggi_3):
  img_4[j, int((lebar_3)/2)] = [255,255,255]

cv2_imshow(img_4)
```

```
# Buat Garis Menyilang
img_cross = cv.cvtColor(image, cv.COLOR_BGR2RGB)

tinggi_cross = img_cross.shape[0]
lebar_cross = img_cross.shape[1]

# Garis dari kiri atas ke kanan bawah
for i in range(min(tinggi_cross, lebar_cross)):
    img_cross[i, i] = [255, 255, 255]

# Garis dari kanan atas ke kiri bawah
for i in range(min(tinggi_cross, lebar_cross)):
    img_cross[i, lebar_cross - 1 - i] = [255, 255, 255]

cv2_imshow(img_cross)
```

```
# Garis Horizontal dengan Lebar Tertentu
img_horizontal = cv.cvtColor(image, cv.COLOR_BGR2RGB)

tinggi_h = img_horizontal.shape[0]
lebar_h = img_horizontal.shape[1]

# Tentukan panjang garis yang diinginkan (dalam piksel)
line_length = 100  # Anda bisa mengubah nilai ini

# Tentukan posisi tengah tinggi gambar
middle_row = int(tinggi_h / 2)

# Tentukan posisi awal dan akhir kolom untuk garis (pusat di tengah gambar)
start_col = int((lebar_h / 2) - (line_length / 2))
end_col = int((lebar_h / 2) + (line_length / 2))

# Pastikan posisi awal dan akhir kolom berada dalam batas gambar
start_col = max(0, start_col)
end_col = min(lebar_h, end_col)

# Gambar garis horizontal
for x in range(start_col, end_col):
    img_horizontal[middle_row, x] = [255, 255, 255]

cv2_imshow(img_horizontal)
```

```
# Buat kotak menggunakan kumpulan pixel warna putih di sembarang tempat dalam gambar
img_box = cv.cvtColor(image, cv.COLOR_BGR2RGB)

tinggi_box = img_box.shape[0]
lebar_box = img_box.shape[1]

# Tentukan posisi dan ukuran kotak
start_x = 50  # Kolom awal
start_y = 50  # Baris awal
box_width = 100 # Lebar kotak
box_height = 80 # Tinggi kotak

# Pastikan koordinat dan ukuran kotak berada dalam batas gambar
end_x = min(lebar_box, start_x + box_width)
end_y = min(tinggi_box, start_y + box_height)

# Gambar kotak
for y in range(start_y, end_y):
    for x in range(start_x, end_x):
        img_box[y, x] = [255, 255, 255]

cv2_imshow(img_box)
```
